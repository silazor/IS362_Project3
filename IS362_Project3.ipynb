```python
import sqlite3 as sql
import os
import pandas as pd

os.chdir("C:/Users/silaz/github/capstone/is362/w9")
dbHandle = sql.connect('file:Chinook_Sqlite.sqlite', uri=True)
cursor = dbHandle.cursor()
sql = '''
    select C.LastName, C.FirstName, T.Name, A.Title
        from Customer C, Track T, Album A, Invoice I, InvoiceLine IL
        WHERE
        C.CustomerId=I.CustomerId and
        I.InvoiceId=IL.InvoiceId and
        IL.TrackId=T.TrackId and
        T.AlbumId = A.AlbumId
        order by C.LastName, C.FirstName;
    '''
df = pd.read_sql_query(sql, dbHandle)
df
```
